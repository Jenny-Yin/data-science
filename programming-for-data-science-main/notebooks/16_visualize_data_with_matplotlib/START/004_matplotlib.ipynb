# Customizing Plots

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Load the Dataset
We will start by importing the necessary libraries and loading the dataset into a pandas DataFrame. We will use the read_csv function to read the data from a CSV file into a DataFrame.

```python
# Load the cereal dataset
df = pd.read_csv("https://raw.githubusercontent.com/cbtn-data-science-ml/python-pandas-for-data-manipulation/main/completed_files/datasets/cereal.csv")
```

```python
df.head()
```

## Create a Basic Scatter Plot
This will create a scatter plot with sugar content on the x-axis and rating on the y-axis.

```python
# Select the sugar and rating columns
x = df['sugars'] 
y = df['rating']

# Create a scatter plot
plt.scatter(x,y)

# Show the plot
plt.show()
```

# Add Axis Labels and Title

```python
# Add axis labels and a title
plt.scatter(x,y)
plt.xlabel('Sugar Content')
plt.ylabel('Rating')
plt.title('Cereal Rating Vs. Sugar Content')
# Show the plot
```

# Change the Marker Size and Color
This will change the size of the markers based on the cereal's calorie content and the color of the markers based on the cereal's fiber content.

```python
# Change the marker size and color
plt.scatter(x,y,s=df['calories'],c=df['fiber'],cmap='Blues')

# Add axis labels and a title
plt.xlabel('Sugar Content')
plt.ylabel('Rating')
plt.title('Cereal Rating Vs. Sugar Content')

# Show the plot
plt.show()
```

# Add a Color Bar
This will add a colorbar to the plot, showing the mapping from fiber content to marker color.

```python
# Change the marker size and color
scatter = plt.scatter(x,y,s=df['calories'],c=df['fiber'],cmap='Blues')

# Add a colorbar
plt.colorbar(scatter)

# Add axis labels and a title
plt.xlabel('Sugar Content')
plt.ylabel('Rating')
plt.title('Cereal Rating Vs. Sugar Content')

# Show the plot
plt.show()
```
