```python
import networkx as nx
```

```python
G = nx.Graph()
```

```python
G.add_nodes_from(["Alice", "Bob", "Charlize", "David", "Eve"])
```

```python
G.add_edge("Alice", "Bob")
G.add_edge("Bob", "Charlize")
G.add_edge("Charlize", "David")
G.add_edge("David", "Eve")
G.add_edge("Eve", "Alice")
```

```python
print(G.edges())
```

```python
import matplotlib.pyplot as plt
```

```python
nx.draw_networkx(G, with_labels=True)

labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_labels(G, pos=nx.spring_layout(G), labels=labels)

pos = nx.spring_layout(G)

node_labels = nx.get_node_attributes(G, 'label')
for node, label in node_labels.items():
        x, y = pos[node]
        plt.text(x, y, label, fontsize=12)

        
plt.show()
```
